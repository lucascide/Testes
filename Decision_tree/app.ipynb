```python
import pandas as pd
import numpy as np

from sklearn import tree
```

```python
estado_civil = [0, 1, 0, 1, 0]
renda = [1000, 2000, 3500, 4000, 5000]
aprovacao = [0,1,0,1,1]

df_aprovacao = pd.DataFrame({'estado_civil':estado_civil, 
                            'renda_mensal': renda, 
                            'aprovado': aprovacao})
```

```python
df_aprovacao
```

```python
X = df_aprovacao.iloc[:, :2]
y = df_aprovacao.iloc[:, -1]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
# Prevê com base na árvore de decisão e retorna um array com a classe prevista
resultado = clf.predict([[1, 1000]])
```
