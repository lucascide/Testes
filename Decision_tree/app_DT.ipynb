```python
import pandas as pd
import numpy as np
from sklearn import tree

path_wine = "C:\\Users\\joaol\\OneDrive\\Documentos\\Codiguinhos\\Python\\Ciência de Dados\\Testes\\datasets-to-study\\wine.csv"
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(path_wine)
df
```

```python
# get the locations
X = df.iloc[:, 1:]
y = df.iloc[:, 0]

# split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
tree.plot_tree(clf)
```
